# Function

```python
import matplotlib.colors as mcolors
from matplotlib.cm import ScalarMappable
from matplotlib.lines import Line2D

def plot_cell2loc_out(
    adata_c2l,
    celltype,
    regions_sel,
    xaxis,
    yaxis,
    abundance_thresh = 0.05,
    mean_abundance_max = 0.3,
    count_log_max=7,
    nspots_for_legend = [0, 50, 100],
    savefig = None
):
    # select spots which have the celltype of interest
    # define based on an abundance threshold
    mask = adata_c2l.to_df()[celltype]>abundance_thresh
    ad = adata_c2l[mask]
    print(f'n_spot which contains {celltype}: {ad.shape[0]}')

    # preparing dataframe
    plot_df = ad.to_df()[[celltype]]
    plot_df[[yaxis,xaxis]] = ad.obs[[yaxis,xaxis]].copy()
    plot_df['count'] = 1
    plot_df = plot_df.groupby([yaxis,xaxis]).sum().loc[regions_sel]
    plot_df['mean'] = plot_df[celltype]/plot_df['count']
    plot_df['count_log'] = np.log1p(plot_df['count'])

    # plotting
    plt.rcParams['axes.grid'] = False
    plt.rcParams['figure.figsize'] = 1.2,4
    norm = mcolors.Normalize(vmin=0, vmax=mean_abundance_max)
    sns.scatterplot(data=plot_df, x=xaxis,y=yaxis,
                    hue='mean', hue_norm=norm,
                    size='count_log', size_norm=(0,count_log_max),sizes=(0, 200),
                   palette='Reds',edgecolor='black')

    # Remove the "mean" legend
    plt.legend([],[], frameon=False)

    # Custom legend for sizes - example
    # Define a few representative sizes and their corresponding original count values
    # sizes = nspots_for_legend
    log_sizes = np.log1p(nspots_for_legend)  # Calculate log sizes for positioning in the plot (optional if you want exact matches)
    # Calculate markersizes based on your specific transformation and plotting parameters
    # This step involves converting log_sizes into the scale used in your plot
    markersizes = np.interp(log_sizes, [0, count_log_max], [0, 200])  # Adjust the range [0, 200] as per your plot
    # Create custom legend handles for sizes
    size_handles = [Line2D([0], [0], marker='o', color='w', markerfacecolor='gray', 
                       markersize=np.sqrt(markersize), label=f'{size}') for size, markersize in zip(nspots_for_legend, markersizes)]

    # Add the custom size legend to the plot
    plt.legend(handles=size_handles, title="n_spot", bbox_to_anchor=(2.5, 1), loc=2, borderaxespad=0.)

    # Show the colorbar with the normalized hue scale
    cbar = plt.colorbar(ScalarMappable(norm=norm, cmap="Reds"))
    cbar.set_label('Mean abundance', rotation=270, labelpad=15)
    plt.xticks(rotation=90)
    plt.margins(x=0.3,y=0.07)
    plt.title(celltype)
    if savefig!=None:
        plt.savefig(savefig, bbox_inches="tight",dpi=300)
    plt.show()
```
