## Create Pseudo-bulk Replicates

```r
proj = addGroupCoverages(ArchRProj = proj, groupBy = "fine_grain", sampleLabels = "Sample2",
                        minReplicates = 3,
                        maxReplicates = 3,
                         minCells = 40,
                      maxCells = 500,
                        force = TRUE)
# maxReplicate = sample number
```

## Estract peakset as Granges object

```r
PeakSet = getPeakSet(proj)
```

```r
# cell number
n_cell = table(proj@cellColData$fine_grain)

# peak number
n_peak = sort(table(rownames(mcols(PeakSet))),decreasing = TRUE)
```

```r
library(ggplot2)
library(ggrepel)

options(repr.plot.width=10, repr.plot.height=10)

df = data.frame(fine_grain=names(n_peak),
           n_peak=as.numeric(unname(n_peak)),
           n_cell=as.numeric(unname(n_cell[names(n_peak)])))
df = df[order(-df$n_peak),]

ggplot(df, aes(n_cell,n_peak)) +
  geom_point()
```

```r
ggplot(df, aes(n_cell,n_peak)) +
  geom_point() +
  geom_text_repel(aes(label = fine_grain))
```
