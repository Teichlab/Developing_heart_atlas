```python
DAR_res_dict={}
for file in os.listdir(dar_res_dir):
    if '_fine-grain_all.csv' in file:
        celltype = file.split('_fine-grain_all.csv')[0]
        # read in 
        df = pd.read_csv(f'{dar_res_dir}/{file}')
        # make peak names
        df['names'] = df['seqnames']+':'+df['start'].astype('str')+'-'+df['end'].astype('str')
        # rename fine_grain celltypes
        df['celltype'] = [celltype_rename[x] for x in df['celltype']]
        # add to the dictionary
        DAR_res_dict[celltype] = df.copy()
DAR_res_dict.keys()
```

# Set celltypes to analyse

```python
import gseapy as gp
from gseapy import barplot, dotplot
```

```python
celltype_reference = 'Cardiomyocytes'
celltype_of_interest = 'SinoatrialNodePacemakerCells'

# select specific and accessible peaks
accessible_proportion_thresh = 0.1
pval_thresh = 0.05
logFC_thresh_specific = 1


### Read in pyscenic region-gene metadata ###

# read in pyscenic outputs
eRegulon_metadata = pd.read_csv(f'/nfs/team205/heart/pyscenic/plus/eRegulon_metadata/{celltype_reference}_eRegulon_metadata_filtered.csv')
eRegulon_metadata['regulon_name_short'] = [x.split('_(')[0][:-2] for x in eRegulon_metadata['Region_signature_name']]
print(eRegulon_metadata.shape)

### Select accessible and specific peaks ###

# select peaks which is differentially accessible
DAR_res = DAR_res_dict[celltype_reference]
DAR_res = DAR_res[DAR_res['celltype']==celltype_of_interest]
DAR_res = DAR_res[(DAR_res['FDR']<pval_thresh)&(DAR_res['Log2FC']>logFC_thresh_specific)]

# select peaks which is accessible more than cetrain proportion of the cell type
acc_mat_bi = (acc_mat > accessible_proportion_thresh).astype('int')
acc_peaks = list(acc_mat_bi[acc_mat_bi[celltype_of_interest]==1].index)
DAR_res = DAR_res[DAR_res['names'].isin(acc_peaks)]

# add peaktype
DAR_res['peakType']=list(adata_peak.var.loc[list(DAR_res['names']),'peakType'])

# get the selected peaks
spec_peaks = list(DAR_res['names'])
print(len(spec_peaks))
print(len(acc_peaks))

### Get genes which linked to the peaks ###
# metadata of regions regulons which contain the selected peaks
eRegulon_metadata_sel = eRegulon_metadata[eRegulon_metadata['Region'].isin(spec_peaks)]
print(eRegulon_metadata_sel.shape)

# genes linked to the selected peaks
# and expressed in the cell type of interst 
expressed_proportion_thresh = 0.1
exp_genes = list(exp_mat.index[exp_mat[celltype_of_interest]>expressed_proportion_thresh])
print(f'n_genes expressed in {celltype_of_interest}: {len(exp_genes)}')

# get linked and expressed genes
df = eRegulon_metadata_sel[eRegulon_metadata_sel['Gene'].isin(exp_genes)]
df = df[['Region','Gene']]
df.drop_duplicates(inplace=True)
linked_genes = list(set(df['Gene']))
rank = df['Gene'].value_counts()
print(f'n_genes linked to specific peaks: {len(linked_genes)}')

# get genes which linked to enhancer peak
df['peakType'] = list(adata_peak.var['peakType'].loc[df['Region']])
linked_genes_enh = list(set(df[df['peakType'].isin(['Distal','Intronic'])]['Gene']))
print(f'n_genes linked to specific peaks (Distal, Intronic): {len(linked_genes_enh)}')

# plot
n_rank = 50
plt.rcParams['axes.grid'] = False
plt.rcParams['figure.figsize'] = 4.5,12
plt.barh(width=rank[:n_rank],y=rank[:n_rank].index,color='darkred')
plt.gca().invert_yaxis()
plt.title(f'Number of {celltype_of_interest} specific peaks')
plt.margins(y=0.01)
plt.show()

### Pathway enrichment analysis of the linked genes ###
# default: Human
geneset_list = ['MSigDB_Hallmark_2020','KEGG_2021_Human','GO_Biological_Process_2023','Reactome_2022']
enr = gp.enrichr(gene_list=linked_genes_enh, # or "./tests/data/gene_list.txt",
                 gene_sets=geneset_list,
                 organism='human', # don't forget to set organism to the one you desired! e.g. Yeast
                 outdir=None, # don't write to disk
                )

geneset = 'GO_Biological_Process_2023'
df = enr.results[enr.results['Gene_set'] == geneset]
df = df[df['Adjusted P-value'] < 0.1]
df = df.sort_values('Combined Score',ascending=False)

# plot
n_rank = 10
plt.rcParams['axes.grid'] = False
plt.rcParams['figure.figsize'] = 6,4
plt.barh(width=df[:n_rank]['Combined Score'],
         y=[x.split(' (')[0] for x in df[:n_rank]['Term']],
         color='darkred')
plt.gca().invert_yaxis()
plt.xlabel('Score')
plt.title(geneset)
plt.margins(y=0.05)
plt.savefig(f'{figdir}/Peak_linked-gene-pathways_{celltype_of_interest}.pdf', bbox_inches="tight",dpi=300)
plt.show()


for geneset in geneset_list:
    print(geneset)
    df = enr.results[enr.results['Gene_set'] == geneset]
    display(df.sort_values('Combined Score',ascending=False)[:20])
```

```python
# proportion of peaktypes
# specific&accessible peak and all peaks
df = pd.DataFrame({
    'type':list(np.repeat('All accessible peaks',len(acc_peaks)))+list(np.repeat('Specific peaks',len(spec_peaks))),
    'peakType':list(adata_peak.var.loc[acc_peaks,'peakType'])+list(adata_peak.var.loc[spec_peaks,'peakType'])
})
df['peakType'] = df['peakType'].astype('category')
df['peakType'].cat.reorder_categories(['Promoter','Exonic','Intronic','Distal'],
                                            inplace=True)

tmp = pd.crosstab(df['type'], df['peakType'], normalize='index')
# plot
plt.rcParams['figure.figsize'] = 2,3
tmp.plot.bar(stacked=True,width=0.7,legend=False)
plt.gca().grid(False)
plt.gca().set_title(celltype_of_interest)
plt.gca().set_xlabel('')
plt.gca().set_xticklabels(plt.gca().get_xticklabels(), rotation=45, 
                          ha='right',rotation_mode='anchor')
plt.gca().set_ylabel('proportion')
plt.gca().legend(bbox_to_anchor=(1.0, 1.0))
plt.show()
```
