# Prepare functions

# Read in anndata

```python
adata_vis = sc.read_h5ad('/nfs/team205/heart/anndata_objects/Foetal/visium-OCT_foetal_raw.h5ad')

# add abundance to .obs
df = adata_vis.obsm['q05_cell_abundance_w_sf'].copy()
## rename columns
df.columns = [x.replace('q05cell_abundance_w_sf_','') for x in df.columns]
adata_vis.obs[df.columns]=df.copy()

# add prop
## convert to proportions # should use np.nansam rather than sum
df = df.apply(lambda x: x/np.nansum(x), axis=1) 
adata_vis.obsm['prop']=df.copy()
adata_vis.obs['prop_'+df.columns] = df.copy()

# set symboßl as var_names
adata_vis.var.set_index('SYMBOL',inplace=True)

# log-normalise
sc.pp.filter_genes(adata_vis, min_cells=3)
sc.pp.normalize_total(adata_vis, target_sum=1e4)
sc.pp.log1p(adata_vis)

adata_vis
```

```python
# add "annotation_final"
annot = pd.read_csv('/nfs/team205/heart/anndata_objects/Foetal/visium_foetal_healthy-all_spot-annotation.csv',
                    index_col=0
                   )
adata_vis.obs['annotation_final'] = annot[['final']].reindex(adata_vis.obs_names)
adata_vis.obs.head()
```
