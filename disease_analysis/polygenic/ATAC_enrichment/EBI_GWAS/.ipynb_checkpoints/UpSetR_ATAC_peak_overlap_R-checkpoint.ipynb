```r
install.packages("UpSetR")
```

```r
library("UpSetR")
library("repr")
library("tidyverse")
```

```r
df=read.csv('/nfs/team205/heart/EBI_GWAS/SNP_mapped_to_peaks/all_peaks_with_SNPs_for_34_traits_incremental.csv')
df=df[, -c(1:5)]
head(df,3)
print(dim(df))
```

```r
snp_md<-read.csv('/nfs/team205/heart/EBI_GWAS/SNP_mapped_to_peaks/SNPs_md.csv')
head(snp_md,5)
print(dim(snp_md))
```

```r
# Replace EFO ID with EFO Term

mm <- match(names(df), snp_md$efo_id)
names(df)[!is.na(mm)] <- as.character(snp_md$efo_term[na.omit(mm)])
head(df,3)
print(dim(df))
```

```r
options(repr.plot.width=20, repr.plot.height=30)
```

```r
upset(df,
     order.by='freq',
      nsets=150,
         text.scale = c(2, 2, 2,1, 1.5, 2),
     mainbar.y.label = "Intersection size",
     sets.x.label = "# open peaks containing GWAS SNPs",
  point.size = 3,
      line.size = 1)
```
