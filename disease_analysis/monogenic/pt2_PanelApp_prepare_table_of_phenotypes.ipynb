# Get ClinVar genes

```python
def getClinVargenes(
    term, # search term for clinvar
    sortby='symbol'
):
    term.strip().replace(" ","+")
    target = f'https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?db=clinvar&term={term}&retmode=json'
    method = 'GET'
    h = http.Http()
    response, content = h.request(target,method)

    clinvarIDs_json = json.loads(content) # parses the JSON
    clinvarIDs_json['esearchresult']

    clinvarIDs = clinvarIDs_json['esearchresult']['idlist']

    term_associated_genes = []
    term_associated_traits = []
    pathogenicity = []
    pathogenicity_evaluation_type = []
    pathogenicity_evaluation_date = []
    variant_type = []
    clinvar_accession = []
    variant_ID = []

    for i in range(len(clinvarIDs)):
        target = f'https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esummary.fcgi?db=clinvar&id={clinvarIDs[i]}&retmode=json'
        method = 'GET'
        h = http.Http()
        response, content = h.request(target,method)
        # parses the JSON
        data2 = json.loads(content)
        term_associated_genes.append(data2['result'][clinvarIDs[i]]['genes'][0]['symbol'])
        traits = []
        for trait in range(len(data2['result'][clinvarIDs[i]]['trait_set'])):
            traits = data2['result'][clinvarIDs[i]]['trait_set'][trait]['trait_name']
        term_associated_traits.append(traits)
        pathogenicity.append(data2['result'][clinvarIDs[i]]['clinical_significance']['description'])
        pathogenicity_evaluation_type.append(data2['result'][clinvarIDs[i]]['clinical_significance']['review_status'])
        pathogenicity_evaluation_date.append(data2['result'][clinvarIDs[i]]['clinical_significance']['last_evaluated'])
        variant_type.append(data2['result'][clinvarIDs[i]]['obj_type'])
        variant_ID.append(data2['result'][clinvarIDs[i]]['variation_set'][0]['variation_name'])
        clinvar_accession.append(data2['result'][clinvarIDs[i]]['accession'])
    clinvar_results = {
        'symbol':term_associated_genes,
        'trait':term_associated_traits,
        'pathogenicity':pathogenicity,
        'pathogenicity_evaluation_type':pathogenicity_evaluation_type,
        'pathogenicity_evaluation_date':pathogenicity_evaluation_date,
        'variant_type':variant_type,
        'variant_ID':variant_ID,
        'clinvar_accession':clinvar_accession,
    }
    clinvar_results=pd.DataFrame.from_dict(clinvar_results).sort_values(sortby,ascending=True)
    print(f'Clinvar search term: "{term}"...')
    return clinvar_results
```
