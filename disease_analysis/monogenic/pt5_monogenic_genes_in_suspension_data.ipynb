# Plot each trait in each cell type

```python
# make a score for each trait 

traits=['ASD','VSD','Obstructive','Malformation of outflow tracts','AVSD','Functional single ventricle','Heterotaxy']

for trait in traits:
    sc.tl.score_genes(adata,
                      gene_list=df.index[(df['CHD classification'].str.contains(trait))&(df['Inheritance mode']=='AD')].tolist(),
                      score_name=trait)
```

```python
sc.pl.matrixplot(adata,
             groupby='mid_grain',
             var_names = traits,
             title='Congenital Heart Disease (CHDgene, AD inheritance only)',
                 standard_scale='var',
                dendrogram=True,
                swap_axes=True,
                colorbar_title='Score',
                cmap='Reds')
```

```python
#Plot all available scores at the gene level
traits=[item for item in traits if item in adata.obs.columns]

for trait in traits:
    trait_genes=df.index[(df['CHD classification'].str.contains(trait))&(df['Inheritance mode']=='AD')].tolist()
    trait_genes=[gene for gene in trait_genes if gene in adata.raw.var_names]
    trait_genes = list(dict.fromkeys(trait_genes)) # removes duplicate items in list
    sc.pl.dotplot(adata,
             groupby='mid_grain',
             var_names = trait_genes,
              standard_scale='var',
                dendrogram=True,
                swap_axes=True,
                title=trait)
```
