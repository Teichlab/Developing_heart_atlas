# Parameters

```python
celltype = 'Cardiomyocytes'
celltype_column = 'coarse_grain'
target_type = 'gene' # gene or region
query_adata_path = '/nfs/team205/heart/anndata_objects/Foetal/trisomy21/AtlasT21IntegratedRaw.h5ad'
path_to_pruned_network = '/nfs/team205/heart/Vitro/TF_selection/foetal_vCM_left_compact-vs-trabeculated_downstream/TF-TG-network_compact-common-up_2Feb2024.pkl'
```

# Read in regulon network

```python
network = utils.read_pkl(path_to_pruned_network)
print(len(network.keys()))
```

# Prepare query data

```python
adata = sc.read_h5ad(query_adata_path)

# subset celltype
adata = adata[adata.obs[celltype_column]==celltype]

# subset multiome
adata = adata[adata.obs['kit_10x']=='Multiome-v1']

# log-normalise
# shouldn't matter whether raw or normalised, need to check
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)

print(adata.X.data[:10])
adata
```

```python
adata.obs['kit_10x'].value_counts()
```

# Score eRegulons

```python
%%time
adata.obsm[f'network_score_{target_type}'] = regulon.score_eRegulons_mod(adata,
                                                network,
                                                # enrichment_type=target_type,
                                                use_rep = 'X',
                                                auc_threshold=0.05,
                                                normalize=False,
                                                n_cpu=1,
                                                seed= 123
                                                 ).reindex(adata.obs_names)
```

# Differential regulons

```python
# make anndata of AUC score
adata_reg = anndata.AnnData(
    X = adata.obsm[f'network_score_{target_type}'],
    obs = adata.obs,
    var = pd.DataFrame(adata.obsm[f'network_score_{target_type}'].columns).set_index(0)
)
adata_reg
```

```python
# modify labels
adata_reg.obs['fine_grain_diagnosis'] = adata_reg.obs['fine_grain_integrated'].astype('str')+'_'+adata_reg.obs['diagnosis'].astype('str')
# modify labels
adata_reg.obs['fine_grain_diagnosis_mod'] = adata_reg.obs['fine_grain_diagnosis'].copy()
labels = []
for label in adata_reg.obs['fine_grain_diagnosis_mod']:
    if label in ['VentricularCardiomyocytesLeftCompact_Atlas','VentricularCardiomyocytesLeftCompact_T21']:
        labels.append(label.replace('_Atlas','_Euploid').replace('VentricularCardiomyocytesLeftCompact','vCM-LeftCompact'))
    elif '_Atlas' in label:
        labels.append('OtherCMs_Euploid')
    elif '_T21' in label:
        labels.append('OtherCMs_T21')
adata_reg.obs['fine_grain_diagnosis_mod'] = labels.copy()
set(adata_reg.obs['fine_grain_diagnosis_mod'])
```

```python
# differential test
# subsetting only Left compact cells and compare between euploid and T21
adata_reg_sub = adata_reg[adata_reg.obs['fine_grain_integrated'].isin(['VentricularCardiomyocytesLeftCompact'])]
print(adata_reg_sub.obs['diagnosis'].value_counts())
sc.tl.rank_genes_groups(adata_reg_sub,groupby='diagnosis',method='t-test')
# get only significant ones
res = sc.get.rank_genes_groups_df(adata_reg_sub,
                            group='T21',
                           pval_cutoff=0.05,
                           )
```

```python
# select regulons
absolute_logFC_thresh = 0.1

T21_down_regulons = list(res[res['logfoldchanges'] <= -absolute_logFC_thresh]['names'])
print(f'T21-down regulon number: {len(T21_down_regulons)}')
T21_up_regulons = list(res[res['logfoldchanges'] >= absolute_logFC_thresh]['names'])
print(f'T21-up regulon number: {len(T21_up_regulons)}')

# plot
sc.pl.dotplot(adata_reg[adata_reg.obs['fine_grain_diagnosis_mod'].isin(['OtherCMs_Euploid',
                                                                        'OtherCMs_T21',
                                                                        'vCM-LeftCompact_Euploid',
                                                                        'vCM-LeftCompact_T21'])],
              T21_down_regulons+T21_up_regulons, # regulons upregulated in Euploid samples
                                groupby='fine_grain_diagnosis_mod',
                                dendrogram=False,
                               standard_scale='var',
              colorbar_title='Mean AUC score\nin group',
              show=False
             )
plt.savefig(f"{figdir}/T21-Euploid-LeftCompact_regulon-score_logFC-tresh-{absolute_logFC_thresh}.pdf", bbox_inches="tight",dpi=300)
```
