```python
sex_genes=sc.get.rank_genes_groups_df(atlas_adata,group='male')
sex_genes=sex_genes.reset_index()
sex_genes=sex_genes.merge(genes.drop_duplicates('gene_name'), left_on='names', right_on='gene_name',how='left')
sex_genes['sex_chromosome']=sex_genes.seq_name.isin(['X','Y'])
sex_genes=sex_genes.set_index('names')
sex_genes.head(3)
```

```python
sex_genes.scores.hist(bins=100)
```

```python
# Calculate the mean and standard deviation
mean_scores = sex_genes['scores'].mean()
std_scores = sex_genes['scores'].std()

# Calculate the upper and lower bounds
lower_bound = mean_scores - 2 * std_scores
upper_bound = mean_scores + 2 * std_scores

# Add a new column 'significant' based on whether scores exceed these bounds
sex_genes['sex_specific'] = (sex_genes['scores'] > upper_bound) | (sex_genes['scores'] < lower_bound)

# Annotate as male or female
sex_genes['sex'] = np.where(sex_genes['scores'] > 0, 'male', 'female')
sex_genes['sex'] = sex_genes['sex'].astype('category')

# Print the bounds for reference
print(f"Lower bound: {lower_bound}")
print(f"Upper bound: {upper_bound}")
```

## Regulon analysis: Compact T21 vs Healthy

```python
dir="/nfs/team205/heart/pyscenic/plus/from_Semih/CardioGRNs/Filtered"
df=pd.read_csv(os.path.join(dir,f"filtered_GRN_for_{list_of_cell_types[0]}.csv"),index_col=0)
df.sort_values("source", ascending=False).head(20)
```

```python
dir="/nfs/team205/heart/pyscenic/plus/from_Semih/CardioGRNs/Raw"
df=pd.read_csv(os.path.join(dir,f"raw_GRN_for_{list_of_cell_types[0]}.csv"),index_col=0)
df.sort_values("source", ascending=False).head(20)
```

```python
TF_dict=df.groupby('source')['target'].apply(list).to_dict()
```

```python
from scipy.stats import fisher_exact
from statsmodels.sandbox.stats.multicomp import multipletests
import numpy as np
import pandas as pd

def get_enriched_TFs(TF_TG_dict, genesets_dict, total_genes_count):
    # Initialize a list for storing TFs and p-values
    results = []

    # Iterate over all genesets
    for geneset_name, geneset in genesets_dict.items():
        # Convert geneset to a set for faster operations
        geneset_set = set(geneset)
        geneset_count = len(geneset_set)

        # Iterate over all TFs
        for tf, genes in TF_TG_dict.items():
            tf_genes_set = set(genes)
            tf_genes_count = len(tf_genes_set)
            
            # Find the intersection with geneset
            overlap_set = geneset_set & tf_genes_set
            overlap_count = len(overlap_set)

            if overlap_count > 0:
                # Perform Fisher's Exact Test
                # Create a contingency table
                contingency_table = np.array([[overlap_count, geneset_count - overlap_count], 
                                              [tf_genes_count - overlap_count, total_genes_count - geneset_count - tf_genes_count + overlap_count]])
                
                # Run Fisher's Exact Test
                _, p_value = fisher_exact(contingency_table, alternative='greater')

                # Store the TF, p_value, and geneset name
                results.append({"TF": tf, "pval": p_value, "geneset": geneset_name})

    # Convert list of dictionaries to DataFrame
    df = pd.DataFrame(results)

    # Apply the Benjamini-Hochberg correction for multiple testing
    reject, pvals_corrected, _, _ = multipletests(df['pval'], method='fdr_bh')
    df['pval_corrected'] = pvals_corrected
    df['reject'] = reject
    df['neglog_pval'] = -np.log10(df['pval_corrected'])

    # Filter the DataFrame for significant results after correction
    df = df[df['reject'] == True]

    return df
```

```python
up = up.dropna()
up = up[up.sex_chromosome==False]

down = down.dropna()
down = down[down.sex_chromosome==False]

genesets_dict={
    'T21':up.index.tolist(),
    'Atlas':down.index.tolist()
}
```

```python
results=get_enriched_TFs(TF_dict,genesets_dict,total_genes_count=36000)
results.sort_values(['geneset','neglog_pval'],ascending=False)
```

```python
plot_data=results.pivot(columns='geneset',index='TF',values='neglog_pval')
plot_data=plot_data.fillna(0)
plot_data=plot_data.reset_index()
plot_data=plot_data.set_index('TF')
plot_data
```

```python
# Abbreviate to top X TFs
AtlasTop=plot_data.sort_values('Atlas',ascending=False).head(20).index.tolist()
AtlaT21Top=plot_data.sort_values('T21',ascending=False).head(20).index.tolist()

# Convert lists to sets
set1 = set(AtlasTop)
set2 = set(AtlaT21Top)

# Find union
union = set1 | set2

# If you want the result as a list
union_list = list(union)

plot_data=plot_data[plot_data.index.isin(union_list)]
plot_data
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap, TwoSlopeNorm
import seaborn as sns

# Define colors for your colormap
colors = ["darkgray", "white", "red"]

# Create a colormap
cmap = LinearSegmentedColormap.from_list("", colors)

# Create the normalization object - white indicates -logp of 1.3 (i.e. adjusted p val>0.05, i.e. non significant)
norm = TwoSlopeNorm(vcenter=1.3)

# Generate clustermap with custom colormap
g = sns.clustermap(plot_data, cmap=cmap, norm=norm, figsize=(1,10), cbar_pos=(2,0.4,0.05,0.05), col_cluster=False, yticklabels=True)

# Add a label to the color bar
g.cax.set_xlabel('-log(p)', rotation=90)

# Set x and y labels
g.ax_heatmap.set_xlabel("Condition")
g.ax_heatmap.set_ylabel("Regulon")

# Add a title above the heatmap
plt.text(0.55, 1.02, "Atlas vs T21: Regulon enrichment in "+list_of_cell_types[0],
         horizontalalignment='center',
         verticalalignment='center',
         transform=g.ax_heatmap.transAxes)

file_path="/lustre/scratch126/cellgen/team205/jc48/jupyter/scArches/results/plots/Atlas21LvCMcompactHeatmap.svg"
plt.savefig(file_path, format='svg', bbox_inches='tight')

plt.show()
```
