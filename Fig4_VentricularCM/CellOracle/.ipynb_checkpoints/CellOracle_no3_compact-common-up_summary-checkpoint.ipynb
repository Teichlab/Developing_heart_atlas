# Plot inner product scores

```python
from scipy.interpolate import UnivariateSpline

nrows = 10
ncols = 7
fig,axs=plt.subplots(nrows=nrows,ncols=ncols,figsize=(25,25)) 
for i,tf in enumerate(dev_dict.keys()):
    dev = dev_dict[tf]
    # get inner product results
    inner_product_df = dev.inner_product_df.copy()
    # normalise pseudotime (min-max)
    arr = np.array(inner_product_df['pseudotime'])
    inner_product_df['pseudotime_norm'] = (arr-np.min(arr))/(np.max(arr)-np.min(arr))
    # sort
    inner_product_df.sort_values('pseudotime_norm',ascending=True,inplace=True)
    # smoothing inner product score
    spl_res = UnivariateSpline(inner_product_df['pseudotime_norm'], inner_product_df['score'], k=2)
    inner_product_df['score_smoothed'] = spl_res(inner_product_df['pseudotime_norm'])
    # plot
    axs[i//ncols,i%ncols].scatter(x=inner_product_df['pseudotime_norm'],y=inner_product_df['score'])
    axs[i//ncols,i%ncols].plot(inner_product_df['pseudotime_norm'],inner_product_df['score_smoothed'],color='red')
    axs[i//ncols,i%ncols].set_title(tf)
    axs[i//ncols,i%ncols].set_xlabel('pseudotime')
    axs[i//ncols,i%ncols].set_ylabel('inner product score')
# fig.suptitle('')
fig.tight_layout()
```

# Project PS score on TF-TF network

```python
# read in TF-TF network
network_TF = utils.read_pkl('/nfs/team205/heart/Vitro/TF_selection/foetal_vCM_left_compact-vs-trabeculated_downstream/TF-TF-network_compact-common-up_2Feb2024.csv')
len(network_TF.keys())
```

```python
# prepare color map
# total PS score

from matplotlib.colors import TwoSlopeNorm

# vector of continuous values
values = np.array(TF_df['total_PS_score'])

# Find the maximum absolute value for normalization
max_abs_value = np.max(np.abs(values))
# Normalize the values manually to ensure zero is centered
# Map the range [-max_abs_value, max_abs_value] to [0, 1] with 0.5 at zero
normalized_values = 0.5 + (values / (2 * max_abs_value))

# Use a diverging colormap
cmap = plt.get_cmap('PiYG')
colors = cmap(normalized_values)

# make mapping dictionary
nodes_color_map = dict(zip(TF_df.index,colors))
```

```python
# prepare parameters
size_scale = 7000

centrality_type = 'eigenvector'
centrality_dict = regulon_plot.get_centrality(network_TF,
                   directed=True,
                   centrality_type=centrality_type, # or 'degree','in_dgree','out_degree'
                   out_df=False # if False output dictionary
                  )

node_size_map = {}
for n in network_TF.keys():
    if n in centrality_dict.keys():
        node_size_map[n]=int(centrality_dict[n]*size_scale)
    else: # means no connection with other TFs
        node_size_map[n]=0
```

```python
# select CHD TFs which are in the network constructed
all_TFs = []
for x in network_TF.values():
    all_TFs = all_TFs + x
all_TFs = all_TFs + list(network_TF.keys())

CHD_TFs_sel = list(set(CHD_TFs).intersection(all_TFs))
print(len(CHD_TFs_sel))
CHD_TFs_sel
```

```python
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable
from matplotlib.colorbar import ColorbarBase
from mpl_toolkits.axes_grid1 import make_axes_locatable

fig, ax = plt.subplots(figsize=(8,8))

regulon_plot.draw_networkx(
    network=network_TF,
    nodes_color_base='white',
    nodes_color_map=nodes_color_map,
    nodes_size_base=400,
    nodes_size_map=node_size_map,
    nodes_edge_highlight=CHD_TFs_sel,
    nodes_edge_color='blue',
    nodes_edge_thickness=3,
     spring_layout_k=1.5,
    spring_layout_scale=1,
    font_size=12,
    pos_seed=3,
    ax=ax
)

# Create an axes on the right side of ax. The width of cax will be 5%
# of ax and the padding between cax and ax will be fixed at 0.05 inch.
divider = make_axes_locatable(ax)
cbar_ax = divider.append_axes("right", size="2%", pad=0.01)
# cbar_ax = fig.add_axes([0.9, 0.1, 0.05, 0.3])

# Create a colorbar with correct scaling
# Setup the normalization and colormap for the colorbar
norm = Normalize(vmin=-max_abs_value, vmax=max_abs_value)
sm = ScalarMappable(norm=norm, cmap=cmap)
sm.set_array([])
cbar = plt.colorbar(sm, cax=cbar_ax, orientation='vertical')
cbar.set_label('Perturbation score (sum)')

fig.suptitle('TF network for compact CM maturation')
fig.tight_layout()
# fig.savefig(f"{figdir}/TF-network_compact-common-up.pdf",bbox_inches="tight",dpi=300)
```
