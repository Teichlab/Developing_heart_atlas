```python
# add age as a continous variable
import re

def age_to_days(age_str):
    match = re.match(r'(\d+)W(\d+)D', age_str)
    if match:
        weeks, days = map(int, match.groups())
        return float(weeks * 7 + days)
    else:
        return None

# Assuming 'adata' is an AnnData object
adata.obs['age_days'] = adata.obs['age'].apply(age_to_days)
adata.obs['age_days']=adata.obs['age_days'].astype(float)
```

```python
adata.obs['diagnosis'] = adata.obs['original_or_new'].map({'original': 'Atlas', 'new': 'T21'})
adata.obs.diagnosis.value_counts()
```

```python
pd.crosstab(adata.obs.diagnosis,adata.obs.age_days)
```

```python
adata.obs['fine_grain'] = adata.obs['fine_grain'].astype('str')
adata.obs['fine_CellTypist_majority_voting'] = adata.obs['fine_CellTypist_majority_voting'].astype('str')
adata.obs['fine_grain_merged'] = adata.obs['fine_grain'].replace('nan', np.nan).fillna(adata.obs['fine_CellTypist_majority_voting'].replace('nan', np.nan))
adata.obs['fine_grain_merged'] = adata.obs['fine_grain_merged'].astype('category')
```

```python
adata.obs['mid_grain'] = adata.obs['mid_grain'].astype('str')
adata.obs['mid_CellTypist_majority_voting'] = adata.obs['mid_CellTypist_majority_voting'].astype('str')
adata.obs['mid_grain_merged'] = adata.obs['mid_grain'].replace('nan', np.nan).fillna(adata.obs['mid_CellTypist_majority_voting'].replace('nan', np.nan))
adata.obs['mid_grain_merged'] = adata.obs['mid_grain_merged'].astype('category')
```

```python
sex_genes=sc.get.rank_genes_groups_df(atlas_adata,group='male')
sex_genes=sex_genes.reset_index()
sex_genes=sex_genes.merge(genes.drop_duplicates('gene_name'), left_on='names', right_on='gene_name',how='left')
sex_genes['sex_chromosome']=sex_genes.seq_name.isin(['X','Y'])
sex_genes=sex_genes.set_index('names')
sex_genes
```

```python
sex_genes.scores.hist(bins=100)
```

```python
# Calculate the mean and standard deviation
mean_scores = sex_genes['scores'].mean()
std_scores = sex_genes['scores'].std()

# Calculate the upper and lower bounds
lower_bound = mean_scores - 2 * std_scores
upper_bound = mean_scores + 2 * std_scores

# Add a new column 'significant' based on whether scores exceed these bounds
sex_genes['sex_specific'] = (sex_genes['scores'] > upper_bound) | (sex_genes['scores'] < lower_bound)

# Annotate as male or female
sex_genes['sex'] = np.where(sex_genes['scores'] > 0, 'male', 'female')
sex_genes['sex'] = sex_genes['sex'].astype('category')

# Print the bounds for reference
print(f"Lower bound: {lower_bound}")
print(f"Upper bound: {upper_bound}")
```

```python
def create_signature(adata, cell_type):
    """
    Create a signature DataFrame from Scanpy rank_genes_groups result.
    
    Parameters:
    adata : AnnData object that has been subjected to differential expression analysis.
    cell_type : String representing the cell type label used in the differential expression analysis.
    
    Returns:
    A pandas DataFrame with two columns: '0' (gene names) and '1' (scores).
    """
    
    # Retrieve the rank genes dataframe
    rank_genes_df = sc.get.rank_genes_groups_df(adata, group=cell_type)
    
    # Create the 'signature' dataframe
    signature = rank_genes_df[['names', 'scores']].copy()
    signature.columns = ['0', '1']
    
    return signature
```

```python
signature = create_signature(tmp_adata, 'VentricularCardiomyocytesLeftCompact_T21')
signature
```
