```python
import pandas as pd
```

```python
chd=pd.read_csv('/home/jovyan/data/CHD_gene_with_phenotype_annotation.csv',index_col='gene')
panel=pd.read_csv('/home/jovyan/data/PanelApp_with_phenotype_annotation.csv',index_col='gene')
clinvar=pd.read_csv('/home/jovyan/data/ClinVar_with_phenotype_annotation.csv',index_col='GeneSymbol')
```

```python
phenotype_col='detailed_CHD_phenotype_string'
chd['phenotype']=chd[phenotype_col]

print(chd.shape)
chd.head(1)
```

```python
phenotype_col='Phenotypes'
panel['phenotype']=panel[phenotype_col]

print(panel.shape)
panel.head(1)
```

```python
phenotype_col='PhenotypeList'
clinvar['phenotype']=clinvar[phenotype_col]

print(clinvar.shape)
clinvar.head(1)
```

```python
df=pd.concat([chd,panel,clinvar])
```

```python
df=df.dropna(axis=1)
```

```python
df.to_csv('/home/jovyan/data/combined_tidied_disease_gene_with_phenotype_annotation.csv')
```
