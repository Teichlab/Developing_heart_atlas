# Function

```python
from scipy.stats import ranksums
from statsmodels.stats.multitest import multipletests

def plot_boxplot(adata, 
                 celltype,
                 score_name,
                 logFC_direction = 'euploid', # euploid or tri21 or both
                 fdr_thresh=0.1,
                 palette = ['gray','lightgray'],
                 savepath = None
                ):
    
    # subset adata
    ad = adata[adata.obs['nhood_annotation']==celltype]
    # target nhoods
    mask_sig = ad.obs['SpatialFDR']<fdr_thresh
    mask_eup = ad.obs['logFC']<0
    mask_t21 = ad.obs['logFC']>0
    nhoods_eup = list(ad.obs[mask_sig&mask_eup].index)
    nhoods_t21 = list(ad.obs[mask_sig&mask_t21].index)
    # add nhood labels
    ad.obs['group_to_compare'] = 'reference'
    ad.obs.loc[nhoods_eup,'group_to_compare'] = 'euploid_nhoods'
    ad.obs.loc[nhoods_t21,'group_to_compare'] = 'tri21_nhoods'
    # subset adata
    if logFC_direction=='both':
        ad.obs['group_to_compare'] = ad.obs['group_to_compare'].astype('category')
        ad.obs['group_to_compare'] = ad.obs['group_to_compare'].cat.reorder_categories(
            ['euploid_nhoods','reference','tri21_nhoods'])
    elif logFC_direction in ['tri21','euploid']:
        ad = ad[ad.obs['group_to_compare'].isin(['reference',f'{logFC_direction}_nhoods'])]
        ad.obs['group_to_compare'] = ad.obs['group_to_compare'].astype('category')
        ad.obs['group_to_compare'] = ad.obs['group_to_compare'].cat.reorder_categories(
            [f'{logFC_direction}_nhoods','reference'])
    else:
        return print('ERROR: logFC_direction is not appropriate')
    
    # prepare datafrane for plotting
    data = ad.obs.copy()
    
    ### plot ###
    split_col = 'group_to_compare'
    # Create a catplot with boxplots for each gene
    ncols=1
    g = sns.catplot(x=split_col, y=score_name, col='nhood_annotation', data=data, kind='box', palette=palette,
                    flierprops=dict(marker=''),
                    col_wrap=ncols, sharey=False, sharex=False, height=5, aspect=0.6)
    # Overlay swarmplot on each Axes
    # We'll use the 'hue' parameter to match the subsets and turn off dodge to align the points with the boxes.
    for celltype, ax in zip(data['nhood_annotation'].unique(), g.axes.flat):
        subset = data[data['nhood_annotation'] == celltype]
        sns.swarmplot(x=split_col, y=score_name, data=subset, color='k', size=2, ax=ax, dodge=False)

    # Update titles to remove "gene ="
    for ax, celltype in zip(g.axes.flat, data['nhood_annotation'].unique()):
        middle_ind = round(len(celltype)/2)
        celltype_title = f'{celltype[:middle_ind]}\n{celltype[middle_ind:]}'
        ax.set_title(celltype_title)

    # Rotate x-tick labels and remove x-axis labels
    for i,ax in enumerate(g.axes.flat):
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")  # Rotate x-tick labels
        # ax.set_xticklabels('')
        ax.set_xlabel('')  # Remove x-axis labels
        if i!=0:
            ax.set_ylabel('')

    # Adjust the spacing of the plots
    g.fig.subplots_adjust(wspace=0.4, hspace=0.4)
    
    # Show the plot
    if savepath!=None:
        plt.savefig(savepath,bbox_inches="tight",dpi=300)
    plt.ylabel(score_name)
    plt.show()

    # statistical test
    if logFC_direction=='both':
        # Wilcoxon Rank-Sum (Mann-Whitney U) Test
        # euploid vs reference
        ranksum_eup_vs_ref = ranksums(data[data['group_to_compare']=='reference'][score_name],
                                      data[data['group_to_compare']=='euploid_nhoods'][score_name])
        # trisomy21 vs reference
        ranksum_t21_vs_ref = ranksums(data[data['group_to_compare']=='reference'][score_name],
                                      data[data['group_to_compare']=='tri21_nhoods'][score_name])
        # trisomy21 vs euploid
        ranksum_t21_vs_eup = ranksums(data[data['group_to_compare']=='euploid_nhoods'][score_name],
                                      data[data['group_to_compare']=='tri21_nhoods'][score_name])
        # multipletest correction
        pvals_adjusted = multipletests([ranksum_eup_vs_ref[1],
                                        ranksum_t21_vs_ref[1],
                                        ranksum_t21_vs_eup[1]],alpha=0.05, method='fdr_bh')[1]
        print(f'adjusted p-values of eup-vs-ref, t21-vs-ref, t21-vs-eup: {pvals_adjusted}')
        
    elif logFC_direction in ['tri21','euploid']:
        print(ranksums(data[data['group_to_compare']=='reference'][score_name],
             data[data['group_to_compare']==f'{logFC_direction}_nhoods'][score_name]))
    
```

# vCM Maturation signature genes

```python
# read in gene signature
path_to_gene_signatures = '/rfs/project/rfs-iCNyzSAaucw/kk837/notebooks/Foetal/vCMLeft/gene_signatures.pkl'
target_signatures = ['compact_specific_up','common_up']

gene_signatures = utils.read_pkl(path_to_gene_signatures)
print(gene_signatures.keys())

sig_genes = []
for sig in target_signatures:
    print(sig)
    sig_genes = sig_genes + gene_signatures[sig]
print(len(sig_genes))
sig_genes[:5]
```

```python
# set some variables
score_name = 'vCMcompact_maturation_score'
celltypes_to_analyse = ['VentricularCardiomyocytesLeftCompact',
                         'VentricularCardiomyocytesLeftTrabeculated',
                         'VentricularCardiomyocytesRightCompact',
                         'VentricularCardiomyocytesRightTrabeculated',
                        'VentricularCardiomyocytesCycling']

if celltypes_to_analyse!=None:
    print('subsetting')
    adata_sub = adata[adata.obs['nhood_annotation'].isin(celltypes_to_analyse)]
else:
    print('just removing "Mixed"')
    adata_sub = adata[adata.obs['nhood_annotation']!='Mixed']
```

```python
# obtain gene score
sc.tl.score_genes(adata_sub, gene_list=sig_genes, 
                  score_name=score_name, random_state=0)
```

```python
# boxplot
plot_boxplot(adata_sub, 
                 celltype='VentricularCardiomyocytesLeftCompact',
                 score_name=score_name,
                 logFC_direction = 'both', # euploid or tri21
                 fdr_thresh=0.1,
                 palette = ['lightblue','lightgray','lightpink'],
                 savepath = f'{figdir}/Tri21_Milo-nhoods_Boxplot_vCMleftcompact_vCM-compaction-maturation.pdf'
                )
```
