# Plot: Euploid nhoods vs reference nhoods

```python
from matplotlib.colors import ListedColormap
```

```python
fdr_thresh = 0.1
geneset = 'MSigDB_Hallmark_2020'
plt_title = f'Trisomy21-nhood vs Euploid-nhood, {geneset}'
significant_only = True
figsize = (8,5.5)

#################################
# prepare dataframes
df = gsea_res_df[gsea_res_df['geneset'] == geneset]
df_nes = df.pivot(columns='Term',index='celltype',values='nes')
df_nes = df_nes.fillna(0)
df_sig = df.pivot(columns='Term',index='celltype',values='fdr')
df_sig = df_sig<fdr_thresh

if significant_only:
    # select pathways which have significant NES at least in one cell type
    pathways_sel = list(df_sig.columns[df_sig.sum()>0])
    df_nes = df_nes[pathways_sel]
    df_sig = df_sig[pathways_sel]

# plotting
plt.rcParams['axes.grid'] = False
# Custom colormap: add grey for masked areas
g = sns.clustermap(df_nes,figsize=figsize,cmap='coolwarm',center=0, # vmax=5,
                  dendrogram_ratio=0.05, cbar_pos=(-0.07, 0.9, 0.02, 0.10)
                  ) # 

# Get the reordered indices for rows and columns
row_order = g.dendrogram_row.reordered_ind
col_order = g.dendrogram_col.reordered_ind

# Reorder the mask based on the clustering result
df_sig_reorder = df_sig.iloc[row_order, :].iloc[:, col_order]

# Iterate over the mask and add asterisks
for i in range(df_nes.shape[0]):
    for j in range(df_nes.shape[1]):
        if df_sig_reorder.iloc[i, j]:
            g.ax_heatmap.text(j + 0.5, i + 0.7, '*', ha='center', va='center', color='black', fontsize=14)
            
# Add a title to the colorbar
colorbar = g.ax_heatmap.collections[0].colorbar
colorbar.set_label('NES', rotation=270, labelpad=15)
plt.title(plt_title)
plt.show()
```

```python
fdr_thresh = 0.1
genesets = ['MSigDB_Hallmark_2020','KEGG_2021_Human']
remove_keywords = ['infection','disease','cancer','cardiomyopathy','shigellosis',
                   'malaria','hepatitis','bacterial','diabetes','atherosclerosis','sclerosis',
                   'carcinogenesis'
                  ]
plt_title = f'Trisomy21-nhood vs Euploid-nhood, {geneset}'
significant_only = True
figsize = (12,6.5)
savepath = f'{figdir}/Tri21_Milo-reassigned-DEGs_T21-vs-EUPnhoods_GSEA_MSigDB-KEGG.pdf'

#################################
# prepare dataframes
df = gsea_res_df.copy()
df = df[df['geneset'].isin(genesets)]
df.replace({'geneset':{
    'MSigDB_Hallmark_2020':'MH',
    'KEGG_2021_Human':'KEGG',
    'GO_Biological_Process_2023':'GOBP',
    'Reactome_2022':'RT'
}},inplace=True)
df['Term'] = df['geneset'].astype('str')+': '+df['Term'].astype('str')
df_nes = df.pivot(columns='Term',index='celltype',values='nes')
df_nes = df_nes.fillna(0)
df_sig = df.pivot(columns='Term',index='celltype',values='fdr')
df_sig = df_sig<fdr_thresh

if significant_only:
    # select pathways which have significant NES at least in one cell type
    pathways_sel = list(df_sig.columns[df_sig.sum()>0])
    df_nes = df_nes[pathways_sel]
    df_sig = df_sig[pathways_sel]

if remove_keywords!=None:
    pathways_sel = [x for x in df_nes.columns if all([xx.lower() not in x.lower() for xx in remove_keywords])]
    df_nes = df_nes[pathways_sel]
    df_sig = df_sig[pathways_sel]

# plotting
plt.rcParams['axes.grid'] = False
# Custom colormap: add grey for masked areas
g = sns.clustermap(df_nes,figsize=figsize,cmap='coolwarm',center=0, # vmax=5,
                  dendrogram_ratio=0.05, cbar_pos=(-0.07, 0.9, 0.02, 0.10)
                  ) # 

# Get the reordered indices for rows and columns
row_order = g.dendrogram_row.reordered_ind
col_order = g.dendrogram_col.reordered_ind

# Reorder the mask based on the clustering result
df_sig_reorder = df_sig.iloc[row_order, :].iloc[:, col_order]

# Iterate over the mask and add asterisks
for i in range(df_nes.shape[0]):
    for j in range(df_nes.shape[1]):
        if df_sig_reorder.iloc[i, j]:
            g.ax_heatmap.text(j + 0.5, i + 0.7, '*', ha='center', va='center', color='black', fontsize=14)
            
# Add a title to the colorbar
colorbar = g.ax_heatmap.collections[0].colorbar
colorbar.set_label('NES', rotation=270, labelpad=15)
plt.title(plt_title)
if savepath!=None:
    plt.savefig(savepath, bbox_inches="tight",dpi=300)
plt.show()
```
