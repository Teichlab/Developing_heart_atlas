### Stats

```python
import pandas as pd
import numpy as np
from scipy.stats import kruskal, mannwhitneyu
import statsmodels.stats.multitest as smm

def two_factor_test(df,
                    factor_A_col,
                    factor_B_col,
                    score
                   ):
    ### 0. prepare levels for the two factors
    levels_A = list(set(df[factor_A_col]))
    levels_A.sort()
    levels_B = list(set(df[factor_B_col]))
    levels_B.sort()
    
    ### 1. Test the main effect of A using a Kruskal–Wallis test
    # First, group the data by factor A (pooling across B)
    grouped_A = [group[score].values for name, group in df.groupby(factor_A_col)]
    kw_A, p_kw_A = kruskal(*grouped_A)
    print(f"Kruskal-Wallis test for the main effect of {factor_A_col}:")
    print(f"Statistic: {kw_A:.4f}, p-value: {p_kw_A:.4f}")
    
    ### 2. Test the effect of B within each level of A
    for a in levels_A:
        print(f"\nTesting the effect of {factor_B_col} within {factor_A_col} = {a}")
        subset = df[df[factor_A_col] == a]

        # Use Kruskal-Wallis test for factor B within this subset.
        grouped_B = [group[score].values for name, group in subset.groupby(factor_B_col)]
        kw_B, p_kw_B = kruskal(*grouped_B)
        print(f"Kruskal-Wallis test for '{factor_B_col}' within '{factor_A_col}' = {a}: Statistic: {kw_B:.4f}, p-value: {p_kw_B:.4f}")

        # Perform pairwise Mann–Whitney U tests between levels of B
        # (only if you wish to run post-hoc pairwise comparisons)
        condition_keys = sorted(subset[factor_B_col].unique())
        comparisons = []
        p_values = []

        for i in range(len(condition_keys)):
            for j in range(i + 1, len(condition_keys)):
                group1 = subset[subset[factor_B_col] == condition_keys[i]][score]
                group2 = subset[subset[factor_B_col] == condition_keys[j]][score]
                stat, p = mannwhitneyu(group1, group2, alternative='two-sided')
                comparisons.append(f"{condition_keys[i]} vs {condition_keys[j]}")
                p_values.append(p)

        # Apply fdr_bh correction for multiple comparisons
        reject, pvals_corrected, _, _ = smm.multipletests(p_values, method='fdr_bh')
        print(f"Post-hoc pairwise comparisons for '{factor_B_col}' within '{factor_A_col}' = {str(a)}:")
        for comp, p_val, rej in zip(comparisons, pvals_corrected, reject):
            print(f"Comparison {comp}: Corrected p = {p_val:.4f}, Significant = {rej}")
            
    ### 3. return the outputs
    return kw_A, p_kw_A
```

```python
# accuracy_df
two_factor_test(accuracy_df,
                    factor_A_col='weight_to_neighbours',
                    factor_B_col='weight_to_edge',
                    score='accuracy'
                   )
```

```python
# accuracy_df
two_factor_test(accuracy_df,
                    factor_A_col='weight_to_neighbours',
                    factor_B_col='weight_to_edge',
                    score='balanced_accuracy'
                   )
```

```python
# f1 score
y_value_list = [x for x in f1score_df.columns if ('f1_class_' in x)&('Adventitia - Coronary' not in x)]

tissues_significant = []
for tissue in y_value_list:
    print(f'########## {tissue} ##########')
    df = f1score_df[[tissue,'weight_to_neighbours','weight_to_edge']]
    df.columns = ['f1_score','weight_to_neighbours','weight_to_edge']
    kw_A, p_kw_A = two_factor_test(df,
                    factor_A_col='weight_to_neighbours',
                    factor_B_col='weight_to_edge',
                    score='f1_score'
                   )
    if p_kw_A<0.05:
        tissues_significant.append(tissue)
    print(f'')
    print(f'')
```

```python
plt.rcParams['axes.grid'] = False
fig,axs=plt.subplots(nrows=3,ncols=6,figsize=(20,8))
# Turn the (3×6) array into a flat list of 18 Axes
axs = axs.flatten()
for i,y_value in enumerate(y_value_list):
    sns.boxplot(data=f1score_df,x='weight_to_neighbours',y=y_value,
                hue='weight_to_edge',palette='Set2',
                ax=axs[i])
    axs[i].set_title(y_value.replace('f1_class_',''))
    axs[i].set_xlabel('neighbour_weight')
    axs[i].set_ylabel('f1 score')
    axs[i].tick_params(axis='x', rotation=90)
    if i==(len(y_value_list)-1):
        axs[i].legend(loc='lower right',title='edge_weight',bbox_to_anchor=(0, -0.9))
    else:
        axs[i].get_legend().remove()
# fig.suptitle('Main Title for the Figure')
# fig.tight_layout(w_pad=1)
fig.subplots_adjust(
    left=0.1, right=0.9,      # margins from figure edge
    bottom=0.1, top=0.9,
    wspace=0.4,               # horizontal space between cols
    hspace=0.6                # vertical space between rows
)
plt.savefig(f'{figdir}/TissueTypist_cross-validation_each-label-f1score.pdf', bbox_inches="tight",dpi=300)
plt.show()
```
