# Calculate axis

```python
# select cells to calculate axis
cells = [
     'SinoatrialNodePacemakerCellsHorn',
    'SinoatrialNodePacemakerCellsHead',
    'SinoatrialNodePacemakerCellsTail',
]
bdata = adata[adata.obs['fine_grain_4Jan2025'].isin(cells)]
obs_df_sel = bdata.obs.copy()
obs_df_sel.rename(columns={'array_row':'y','array_col':'x'},inplace=True)
```

```python
# calculate distance
print('calculating distances')
dist2clusterall = tt.dist2cluster_fast(obs_df_sel, annotation='SANaxis_landmarks', KNN=1, logscale=False)
dist2clusterall
```

```python
obs_df_sel = tt.calculate_axis_2p(obs_df_sel, 'SANaxis_landmarks', ['start','end'], output_col='out',prefix='L2_dist_')
```

```python
bdata.obs['SAN_axis'] = np.nan
bdata.obs.loc[obs_df_sel.index,'SAN_axis'] = obs_df_sel['out']
```

```python
# add to adata
adata.obs['SAN_axis'] = np.nan
adata.obs.loc[bdata.obs_names,'SAN_axis'] = bdata.obs['SAN_axis'].copy()
```

```python
mask = ((adata.obs['array_row'] >= 1800) & 
        (adata.obs['array_row'] <= 2300) & 
        (adata.obs['array_col'] >= 600) & 
        (adata.obs['array_col'] <= 1000)
       )
bdata = adata[mask]

sc.pl.spatial(bdata, color='SAN_axis',cmap='gist_rainbow',
              img_key="0.3_mpp_150_buffer", basis="spatial_cropped_150_buffer",
              size=7,legend_fontsize=30, spot_size=10,
                  frameon=False) # save=f'_{c}_b2c',
```
